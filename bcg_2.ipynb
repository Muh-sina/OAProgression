```python
# Step 1: Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Load Data
# Upload files manually in Colab or use paths from Google Drive if needed
client_df = pd.read_csv("/content/client_data (1).csv")
price_df = pd.read_csv("/content/price_data (1).csv")

# Step 3: Review Data - Check column names, data types, and missing values
print("Client Data Info:\n", client_df.info(), "\n")
print("Price Data Info:\n", price_df.info(), "\n")

print("Client Data Types:\n", client_df.dtypes, "\n")
print("Price Data Types:\n", price_df.dtypes, "\n")

# Step 4: Summarize Data - Show descriptive statistics
print("Client Data Statistics:\n", client_df.describe(), "\n")
print("Price Data Statistics:\n", price_df.describe(), "\n")

# Step 5: Identify Unique Values in Categorical Columns
for col in ['channel_sales', 'origin_up', 'has_gas']:
    print(f"Unique values in {col}:", client_df[col].unique())

# Step 6: Visualizing Data Distributions
def plot_distributions(df, cols, title):
    num_cols = len(cols)
    rows = (num_cols // 2) + (num_cols % 2)

    plt.figure(figsize=(12, 6))
    for i, col in enumerate(cols):
        plt.subplot(rows, 2, i + 1)
        sns.histplot(df[col], kde=True, bins=30)
        plt.title(col)

    plt.suptitle(title, fontsize=14)
    plt.tight_layout()
    plt.show()

# Select numerical columns to visualize
client_numeric_cols = ['cons_12m', 'forecast_cons_12m', 'net_margin', 'pow_max', 'churn']
price_numeric_cols = ['price_off_peak_var', 'price_peak_var', 'price_off_peak_fix', 'price_peak_fix']

# Step 7: Plot Distributions
plot_distributions(client_df, client_numeric_cols, "Client Data Distributions")
plot_distributions(price_df, price_numeric_cols, "Price Data Distributions")
```
